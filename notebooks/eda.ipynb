## Pengecekan Nilai Outlier

```python
def count_outliers_iqr(df, multiplier=1.5):
    outlier_counts = {}

    for col in df.select_dtypes(include=['int64', 'float64']).columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - multiplier * IQR
        upper_bound = Q3 + multiplier * IQR

        count = ((df[col] < lower_bound) | (df[col] > upper_bound)).sum()
        outlier_counts[col] = count

    return outlier_counts

outlier_counts = count_outliers_iqr(df[['temperature_2m_mean', 'temperature_2m_max', 'temperature_2m_min',
       'apparent_temperature_mean', 'apparent_temperature_max',
       'apparent_temperature_min', 'wind_speed_10m_max', 'wind_gusts_10m_max',
       'wind_direction_10m_dominant', 'shortwave_radiation_sum',
       'et0_fao_evapotranspiration', 'sunrise', 'sunset', 'daylight_duration',
       'sunshine_duration', 'precipitation_sum', 'rain_sum', 'snowfall_sum',
       'precipitation_hours']])

for col, count in outlier_counts.items():
    print(f"Kolom {col}: {count} outlier")
```

## Pengecekan Domain Outlier

```python
thresholds = {
    'temperature_2m_max': (-89.2, 56.7),
    'temperature_2m_min': (-89.2, 56.7),
    'apparent_temperature_max': (-89.2, 56.7),
    'apparent_temperature_min': (-89.2, 56.7),
    'wind_speed_10m_max': (0, 231),
    'wind_gusts_10m_max': (0, 231),
    'wind_direction_10m_dominant': (0, 360),
    'shortwave_radiation_sum': (0, 1400),
    'sunshine_duration': (0, 86400),
    'daylight_duration': (0, 86400),
    'precipitation_sum': (0, 1825),
    'rain_sum': (0, 1825),
    'snowfall_sum': (0, 1000),
    'et0_fao_evapotranspiration': (0, 20),
    'jarak_ke_beban': (0, 100)
}

outlier_errors = {}
for feature, (lower, upper) in thresholds.items():
    mask_outlier = (df[feature] < lower) | (df[feature] > upper)
    outlier_errors[f'outlier_{feature}'] = df[mask_outlier]

print("\nSummary Deteksi Outlier Berdasarkan Threshold:")
for check, data in outlier_errors.items():
    print(f"{check}: {len(data)} kasus outlier")
```

## Validasi Data

```python
#1. Validasi
validation_errors = {}

# 1. Sunshine Duration <= Daylight Duration
mask_sunshine = df['sunshine_duration'] > df['daylight_duration']
validation_errors['sunshine_vs_daylight'] = df[mask_sunshine]

# 2. Temperature Min <= Temperature Max
mask_temp = df['temperature_2m_min'] > df['temperature_2m_max']
validation_errors['temperature_min_max'] = df[mask_temp]

# 3. Apparent Temperature Min <= Apparent Temperature Max
mask_apparent_temp = df['apparent_temperature_min'] > df['apparent_temperature_max']
validation_errors['apparent_temperature_min_max'] = df[mask_apparent_temp]

# 4. Wind Speed <= Wind Gust
mask_wind = df['wind_speed_10m_max'] > df['wind_gusts_10m_max']
validation_errors['wind_speed_gusts'] = df[mask_wind]

# 5. Precipitation Sum, Rain Sum, Snowfall Sum >= 0
mask_precip = (df['precipitation_sum'] < 0) | (df['rain_sum'] < 0) | (df['snowfall_sum'] < 0)
validation_errors['precipitation_negative'] = df[mask_precip]

# 6. Shortwave Radiation Sum >= 0
mask_radiation = df['shortwave_radiation_sum'] < 0
validation_errors['shortwave_radiation_negative'] = df[mask_radiation]

# 7. Sunrise < Sunset
mask_sunrise_sunset = df['sunrise'] >= df['sunset']
validation_errors['sunrise_after_sunset'] = df[mask_sunrise_sunset]

# 8. ET0 FAO Evapotranspiration >= 0
mask_et0 = df['et0_fao_evapotranspiration'] < 0
validation_errors['et0_negative'] = df[mask_et0]

# Summary jumlah error per cek
print("Summary Validasi Fisik Cuaca:")
for check, data in validation_errors.items():
    print(f"{check}: {len(data)} kasus error")
```
